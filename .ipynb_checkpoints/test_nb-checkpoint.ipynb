```python
from classes import User, Listing, Marketplace, Purchases, sigmoid
import numpy as np
import pandas as pd
from scipy.special import expit
```

```python
engagement_heterogeneity_arg_dict = {'shape': .2, 'scale': 5}
listing_quality_arg_dict = {'shape': .2, 'scale': 5}

market = Marketplace(engagement_heterogeneity=np.random.gamma,
                     engagement_heterogeneity_arg_dict=engagement_heterogeneity_arg_dict,
                     listing_quality=np.random.gamma,
                     listing_quality_arg_dict=listing_quality_arg_dict)
```

```python
market.reset()
n_users = 500_000
n_listings = n_users * 2
market.init(n_users=n_users, n_listings=n_listings)
```

```python
p_trx_list = []
sold_ads = []

for day in range(14):
    for uidx, user in enumerate(market.users):
        visits = np.random.binomial(1, expit(-2.5 + .05 * user.engagement)) 

        if visits == 0:
            continue
            
        market.users[uidx].visits += 1
        
        if user.variant is None:
            market.users[uidx].variant = np.random.binomial(1, p=.5)
        
        seller = np.random.binomial(1, expit(-2.5 + .0075 * user.engagement))
        
        # selling
        if seller:
            q = np.random.gamma(**listing_quality_arg_dict)  # new listing quality
            new_listing = user.place_listing(q)
            market.listings.append(new_listing) # put listing in market

        for lidx, listing in enumerate(market.listings):
            # viewing
            p_view = expit(-2 + .05 * listing.quality)
            presented = np.random.binomial(1, p_view)  # listings are not always displayed to every user

            if presented == 0:  
                continue
                
            # market.listings[lidx].views += 1
            user.view(listing)
            
            p_trx =  expit( -5 +
                            0.5  * user.variant +
                           .00005 * listing.quality +
                           .00005  * user.engagement +
                           .00005 * listing.quality * user.engagement)

            do_transact = np.random.binomial(1, p=p_trx)
            
            p_trx_list.append({'user': user.id, 'listing': listing.id, 'view': presented, 'p_trx': p_trx, 'is_trx': do_transact})

            # buying
            if do_transact == 0:  # is transacted, listings goes off market and buyers acquires it
                continue
                
            market.listings[lidx].transactions += 1
            bought = market.listings.pop(lidx)
            sold_ads.append(bought)
            user.buy(bought)
```

```python
pd.DataFrame(p_trx_list).to_csv('events.csv', index=False)
```

```python
users_summary = []

for user in market.users:
    userdict = {'entity': user.id,
                'variant': user.variant,
                'visits': user.visits,
                'listings_placed': len(user.listings),
                'transactions': len(user.purchases)}
    
    users_summary.append(userdict)
```

```python
listing_summary = []
for listing in market.listings:
    listingdict = {'entity': listing.id,
                'quality': listing.quality,
                'views': listing.views,
                'transactions': listing.transactions}
    listing_summary.append(listingdict)
```

```python
df_users = pd.DataFrame(users_summary)
df_listing = pd.DataFrame(listing_summary)
```

```python
df_listing.to_csv('listings_views.csv', index=False)
```

```python
from fisher.statistics import permutation_test
```

```python
out = permutation_test(
    x=df_users.loc[df_users.variant == 0, 'transactions'].values,
    y=df_users.loc[df_users.variant == 1, 'transactions'].values,
    func=lambda x, y: y.mean() / x.mean() - 1,
    paired=False,return_permutations=True
)
```

```python
pd.Series(out[1]).to_csv('permutations_effect.csv', index=False)
```
