algorithm
for every day till end of simulation:
1. all users go through decision: visit or not - it's a function of their engagement (always visit if there is a seller-accepted proposal waiting)
2. **Seller mode.** 
    - the users that decided to visit decide now:
        - users check in box and evaluate proposals
        - to list or not (everyone can be a seller at any given time)
3. **Buyer mode.**  
    - check inbox for accepted proposals and transact with some high likelihood
    - decide to search, if searched, then:
        -  market presents them with a matching listing set
        - for each listing in this set:
                -  decide to view, if view then
                        - decide to reply to seller, if so then
                                -  decide to make a proposal
                                        - 
### assumptions
1. buyers non-transacting interactions with listings is independent of past interactions of their own -- no auto-regulating buying behaviour
2. buyer (non)transacting interactions are dependent of other buyers' transacting interactions on the same listing (set) -- marketplace interference
3. Free market: listings price are not determined by the market "hand", it's determined by supply and demand

```python
import numpy as np
from scipy.stats import gamma
from classes import Marketplace, Category
```

```python
market = Marketplace(engagement_heterogeneity=gamma(a=2, scale=7 / 2),
                     response_heterogeneity=gamma(a=2, scale=1 / 2),
                     name="olx")
```

```python
market.reset()
n_users = 50000
market.initialise(n_users=n_users, listing_kwargs={'category': Category.ELECTRONICS})
# market.run_n_day(1)
```

```python
from cProfile import Profile
with Profile() as pr:
    market.run_n_day(1)
```
